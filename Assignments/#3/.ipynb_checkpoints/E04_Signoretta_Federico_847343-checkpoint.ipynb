**Descrizione**:  implementare uno script di image stitching tra due immagini, partendo dalle istruzioni viste ad esercitazione.

### 1. Import dei pacchetti

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
```

### 2. Caricamento immagini

```python
# Caricamento immagini
img1 = cv.imread('images/boat1.jpg',0)
img2 = cv.imread('images/boat3.jpg',0)

fig = plt.figure(figsize=(18,9))

fig.add_subplot(1,2,1)
plt.imshow(img1, cmap='gray')
plt.title('Immagine 1')
#plt.axis('off')

fig.add_subplot(1,2,2)
plt.imshow(img2, cmap='gray')
plt.title('Immagine 2')
#plt.axis('off')
plt.show()
```

```python
print("Dimensione immagine 1:", img1.shape)
print("Dimensione immagine 2:", img2.shape)
```

### 3. Estrazione keypoint e descrittori
Viene inizializzata l'algoritmo *SIFT_create()*, la quale consentir di individuare i keypoints ed i relativi descrittori delle immagini.

```python
sift = cv.xfeatures2d.SIFT_create()
kps1, features1 = sift.detectAndCompute(img1, None)
kps2, features2 = sift.detectAndCompute(img2, None)
```

```python
print("Numero di keypoint nell'immagine 1:", len(kps1))
print("Numero di keypoint nell'immagine 2:", len(kps2))
```

#### 3.1 Visualizzazione dei keypoint individuati

```python
# Visualizzazione punti SIFT
box_kps_1 = cv.drawKeypoints(img1, kps1, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
box_kps_2 = cv.drawKeypoints(img2, kps2, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

fig = plt.figure(figsize=(18,9))

fig.add_subplot(1,2,1)
plt.imshow(box_kps_1, cmap='gray')
plt.title('Immagine 1')
#plt.axis('off')

fig.add_subplot(1,2,2)
plt.imshow(box_kps_2, cmap='gray')
plt.title('Immagine 2')
#plt.axis('off')
plt.show()
```

### 4. Matching
Ottenuti i descrittori (*features1* e *features2*) ed i keypoint (*kps1* e *kps2*) per entrambe le immagini,  possibile andare a cercare le corrispondenza tra di esse. Quindi, per unire le due immagini,  necessario trovare i punti di overlapping, i quali daranno un'idea dell'orientamento della seconda immagine rispetto alla prima. Per farlo verr utilizzata l'algoritmo *BFMatcher()* con *crossCheck=True*. Il *crossCheck*, permette di restituire solo i valori accoppiati *(i,j)* tali che l'*i-esimo* descrittore dell'immagine1 abbia il *j-esimo* descrittore dell'immagine2 come miglior matching, e viceversa. Questo processo permette di avere un risultato consistente.

```python
# Match features
bf = cv.BFMatcher(crossCheck=True)
matches = bf.match(features1, features2)
```

#### 3.4 Visualizzazione dei migliori match
Vengono visualizzati i migliori 20 match

```python
# Visualizzazione match filtrati
matches = sorted(matches, key=lambda x:x.distance)
result = cv.drawMatches(img1, kps1, img2, kps2, matches[0:20], None, matchColor=[255,255,0])

plt.figure(figsize=(18,9))
plt.imshow(result)
#plt.axis('off')
plt.show()
```

5. Homography
Una volta trovati i migliori punti di corrispondenza tra le immagini, viene calcolata la loro *homography matrix*. La matrice omografica  $H$ ,  una matrice di dimensione  $3$x$3$ , che determina una trasformazione in grado di mappare i punti dell'immagine sorgente (immagine 2),  $x=(x_1,x_2,1)$ , sull'immagine di destinazione (immagine 1),  $x=(x_1',x_2',x_3')$ , attraverso l'equazione $$x=Hx$$
 
Per tale compito verr utilizzato *findHomography()*, servito sempre da *openCV*.

```python
source_pts = np.asarray([kps1[m.queryIdx].pt for m in matches])
destination_pts = np.asarray([kps2[m.trainIdx].pt for m in matches])

H, mask = cv.findHomography(destination_pts, source_pts, cv.RANSAC)
```

#### 5.1 Rappresentazione grafica dell'immagine 2, trasformata mediante la matrice H

```python
warp_imgs = cv.warpPerspective(img2, H, dsize=(1200, 518))
plt.figure(figsize=(16,8))
plt.imshow(warp_imgs, cmap='gray')
#plt.axis('off')
plt.show()
```

### 6. Integrazione tra l'immagine 1 e l'immagine trasformata
Come ultimo step dell'*image stitching*, vengono "cucite" l'immagine 1 ed un ritaglio dell'immagine trasformata.

```python
stitched_img = np.concatenate((img1,warp_imgs[:,img1.shape[1]+1:]), axis=1)
```

```python
plt.figure(figsize=(18,9))
plt.imshow(stitched_img, cmap='gray')
#plt.axis('off')
plt.show()
```

Come  possibile osservare, il risultato ottenuto  abbastanza soddisfacente: l'immagine trasformata sembra essere stata proiettata correttamente, sebbene rimanga evidente la divisione tra le due immagini.
