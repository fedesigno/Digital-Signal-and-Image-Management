### 1. Import dei pacchetti necessari

```python
import sounddevice as sd
from scipy.io.wavfile import write
import speech_recognition as sr
from gtts import gTTS
import os
import time
import playsound
import numpy as np
import librosa
```

### 2. Definizione delle funzioni
- manipolazione delle tracce audio attraverso *add_noise* ape raggiungere rumore, *change_pitch* per cambiare il tono di voce e *change_speed* per velocizzare l'audio
- *recording_audio*: per registrare e salvare nelle cartelle le tracce

```python
def add_noise(data, noise_factor=0.005):
    noise = np.random.randn(len(data)).reshape(-1,1)
    noised_data = (data + noise_factor*noise).astype('float32')
    return noised_data

def change_pitch(data, sampling_rate=88200, pitch_factor=-.8):
    data = data.reshape(-1)
    return librosa.effects.pitch_shift(data, sampling_rate, pitch_factor).reshape(-1,1)

def change_speed(data, speed_factor=1.5):
    data = data.reshape(-1)
    return librosa.effects.time_stretch(data, speed_factor).reshape(-1,1)
```

```python
def recording_audio(name, command):
    
    fs = 88200
    seconds=3
    count=0
    
    for i in range(50):
        fold = name+'_'+command
        if not os.path.exists(fold):
            os.mkdir(fold)
        
        print('Start recording ', str(i+1)+'/50')
        myrecording = sd.rec(int(seconds * fs), samplerate=fs, channels=1)
        sd.wait()  # Wait until recording is finished
        
        myrecording_noise = add_noise(myrecording)
        myrecording_pitch = change_pitch(myrecording)
        myrecording_speed = change_speed(myrecording)
        
        records = [myrecording, myrecording_noise, myrecording_pitch, myrecording_speed]
        
        for r in records:
            write(fold+'/'+fold+'_'+str(count)+'.wav', fs, r)
            count+=1
        print('Files saved!')
        
        time.sleep(2)
```
